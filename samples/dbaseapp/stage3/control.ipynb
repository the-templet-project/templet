**Алгоритм запуска программы для работы со списком пользователей.**
1. Скопируйте **cd ~/templet/samples/dbaseapp/stage3; ./user-list** в буфер обмена выделив текст, затем нажав **\<Ctrl\>+C**.
2. Запустите терминал, нажав на  <button data-commandLinker-command="terminal:create-new">кнопку</button>, затем снова переключитесь в это окно.
3. Вставьте скопированную команду в окно терминала, нажав на <button data-commandLinker-command="terminal:paste">кнопку</button>.
4. Перейдите в окно терминала и запустите скопированную команду, нажав **\<Enter\>**.
