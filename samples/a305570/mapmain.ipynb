```c++
{   
    const int SIZE = 10;
    bool master = true;
    
    mapper::engine eng;
    dls7_mapper a_mapper(eng);
    
    if(master){
        a_mapper.N.resize(SIZE);
        for(int i=0;i<SIZE;i++) a_mapper.N[i] = i;
        a_mapper.init(SIZE);
    }
    
    a_mapper.map();
   
    if(master){
        for(int i=0;i<SIZE;i++)
            std::cout << a_mapper.N[i] <<"^2 = " << a_mapper.NxN[i] << std::endl;
    }
}
```
